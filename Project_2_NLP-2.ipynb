**FEATURE 1: Named Entity Recognition (NER)**


```python
!pip install spacy
!python -m spacy download en_core_web_sm
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
def extract_entities(text):
    doc = nlp(text)

    entities = []

    for ent in doc.ents:
        entities.append((ent.text, ent.label_))

    return entities
```

```python
test_text = "Google developed an AI system in the United States in 2025."

entities = extract_entities(test_text)

print("Extracted Entities:")

for entity, label in entities:
    print(entity, "->", label)
```

**Input for testing**

```python
text = input("Enter research text: ")

entities = extract_entities(text)

print("\nExtracted Entities:")

if len(entities) == 0:
    print("No entities found")

else:
    for entity, label in entities:
        print(entity, "->", label)
```

**FEATURE 2: Topic Classification**

```python
from transformers import pipeline
```

```python
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
topics = [
    "Artificial Intelligence",
    "Machine Learning",
    "Natural Language Processing",
    "Computer Vision",
    "Medical Research",
    "Robotics",
    "Cyber Security",
    "Data Science"
]
```

```python
def classify_topic(text):

    result = classifier(
        text,
        candidate_labels=topics
    )

    return result["labels"][0]
```

```python
test_text = "Deep learning models are used to detect brain tumors from MRI medical images."

topic = classify_topic(test_text)

print("Predicted Research Topic:", topic)
```

Input from user


```python
text = input("Enter research text: ")

topic = classify_topic(text)

print("\nPredicted Research Topic:", topic)
```

**FEATURE 3: Question Answering**

```python
from transformers import pipeline

qa_model = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)
```

```python
def ask_paper(question, text):

    result = qa_model(
        question=question,
        context=text
    )

    return result["answer"]
```

```python
test_text = """
This research uses convolutional neural networks to detect brain tumors
from MRI images. The model improves tumor detection accuracy.
"""

question = "What model is used?"

answer = ask_paper(question, test_text)

print("Question:", question)
print("Answer:", answer)
```

```python
text = input("Enter research paper text: ")

question = input("Ask a question about the research paper: ")

answer = ask_paper(question, text)

print("\nAnswer:", answer)
```

**FEATURE 4: Text Summarization**

```python
from transformers import pipeline
```

```python
summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)
```

```python
def generate_summary(text):

    result = summarizer(
        text,
        max_length=100,
        min_length=30,
        do_sample=False
    )

    return result[0]["summary_text"]
```

```python
test_text = """
Artificial intelligence is rapidly transforming the healthcare industry.
Deep learning models are being used to analyze medical images and detect
diseases such as brain tumors and cancer. These models can process large
amounts of medical data and help doctors make faster and more accurate
decisions. This research focuses on using convolutional neural networks
for medical image analysis and improving disease detection accuracy.
"""

summary = generate_summary(test_text)

print("Original Text:")
print(test_text)

print("\nSummary:")
print(summary)
```

Input from user

```python
text = input("Enter research text: ")

summary = generate_summary(text)

print("\nGenerated Summary:")
print(summary)
```
